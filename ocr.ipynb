```python
import cv2
import numpy as np

#  opencvlogo.png
img = cv2.imread("opencvlogo.png",1)

#  hsv
hsv = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)

# 
lower_blue =  np.array([100,50,50])
higher_blue = np.array([140,255,255])

#
mask = cv2.inRange(hsv,lower_blue,higher_blue)

#
left_blue = cv2.bitwise_and(img,img,mask=mask)

# hsv
cv2.imshow("P3-OpenCV logo",img)
# Blue
cv2.imshow("P3-Blue",left_blue)


cv2.waitKey(0)
```

```python
import cv2
import numpy as np

#  p9.png
org = cv2.imread("map.jpg",1)
img = cv2.imread("map.jpg",1)
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
# 
edges = cv2.Canny(gray,30,250,apertureSize=3)
# 
lines = cv2.HoughLines(edges,1,np.pi/180,200)

for line in lines:
    for rho,theta in line:
        a = np.cos(theta)
        b = np.sin(theta)
        x0 = a*rho
        y0 = b*rho
        x1 = int(x0 + 1000*(-b))
        y1 = int(y0 + 1000*(a))
        x2 = int(x0 - 1000*(-b))
        y2 = int(y0 - 1000*(a))
        #
        cv2.line(img,(x1,y1),(x2,y2),(0,0,255),2)

# 
cv2.imshow("org",org)
cv2.imshow("processed",img) 

cv2.waitKey(0)
```
